In this problem you'll analyze a dataset of the monthly number of passengers on US flights from the years 1949 - 1960.

```python
# Part 1 Solution

# --- write code here ---
flightscsv=pd.read_csv('data/flights.csv')
flights=pd.DataFrame(data=flightscsv)
flights.iloc[0:10]
```

**Part 2** (12 points)  
Plot a histogram of the number of montly passengers traveling during the summer months. Summer months include June, July, and August. Does the histogram resemble a normal distribution? Provide two justifications for your answer.

```python
# Part 2 Solution

# --- write code here ---
summerMonths=['June','July','August']
summerFlights=flights[flights.month.isin(summerMonths)]
plot=summerFlights['passengers'].plot.hist()
plt.xlabel("Passengers")
plt.title("Frequency of Number of Flight Passengers During Summer Months")

# --- written response here ---
# The histogram does not resemble a normal distribution. The mean looks to be in the range of 200 to 300 meanwhile
#the median appears to be more like 400. In a normal distribution the mean and median would be the same. 
# Another reason the plot does not look normally distributed is because the data is not evenly distributed around the mean.
#In a normal plot you would see the same frequency to the left of the mean as you would the right. Here the frequency 
#trends downwards as you move up the x axis. 
```
